## Задача

1. Объявить пустой список (группа студентов, сотрудники, и т.д.)
2. В список добавляется информация о человеке: имя, фамилия, возраст. Каждый объект списка представляет собой словарь: { "name": "Ivan", "surname":  "Ivanov", "age": 18, "number": 1}.
3. Реализовать следующие функции:

1) Добавление человека в список: данные вводятся через консоль в удобном для вас виде, проверка на уникальность номера

2) Вывод ФИ и возраста человека по введенному в консоли номеру в списке в формате: Ivanov Ivan, 18

3) Вывод всех людей с возрастом равным n (возраст вводится в консоли)

4) При добавлении человека информация о нем должна построчно записываться в текстовый файл (дополнить функцию №1)

5) Прочитать все содержимое файла и вывести на экран в формате Ivanov Ivan, 18, отсортировано в алфавитном порядке

6) Реализовать функцию удаления человека по номеру
